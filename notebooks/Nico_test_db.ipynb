**Transformation de la DB des urls en dataframe pandas**

```python
import sqlite3
import pandas as pd


db_path = "C:\\Users\\nicol\\Getting in Trump's Head\\Getting-in-Trump-s-Head\\data\\speeches.db"
conn=sqlite3.connect(db_path)
df1=pd.read_sql_query("Select * from Speeches",conn)
conn.close()
```

1) Combien de secondes de speech de Trump avons-nous en DB?

```python
nbr_seconds=df1[df1['nbr_seconds'].apply(type)==int].nbr_seconds.sum()
nbr_minutes=nbr_seconds//60
nbr_hours=nbr_minutes//60
nbr_days=nbr_hours//24
print(f'La DB contient un total de {nbr_seconds} secondes de Trump qui parle, soit {nbr_minutes} minutes, soit {nbr_hours} heures, ou encore {nbr_days} jours de paroles en continues')
```
